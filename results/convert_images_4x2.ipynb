```python
# code for displaying multiple images in one figure

#import libraries
import os
import cv2
from matplotlib import pyplot as plt
import os
import glob

## FOR 4x2 COMBINED IMAGE GENERATION TEST OUTPUTS

fig = plt.figure(figsize=(10, 5)) # for 4x2

# setting values to rows and column variables
rows = 2
columns = 4
# filelist=os.listdir('hyper_set_7/test_results/')
file_list = glob.glob("hyper_set_7/test_results//*")
n = 1
# print(file_list)
for file in file_list: # filelist[:] makes a copy of filelist.
    if file.endswith('.png'):
        Image = cv2.imread(file)
        fig.add_subplot(rows, columns, n)
        plt.imshow(Image)
        plt.axis('off')
        name='Test Output #'+str(n)
        plt.title(name)
        n+=1

fig.savefig('hyper_set_7/test_results/combined_test.png')
```
